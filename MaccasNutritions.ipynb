```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv")
# set width of bar 
barWidth = 0.25
fig = plt.subplots(figsize =(17, 8)) 
 
# set height of bar 
Calories = [1880, 1150, 1090, 1050, 990] 
CaloriesFat = [1060, 540, 530, 510, 470] 
TotalFat = [118, 60, 59, 56, 52] 
 
# Set position of bar on X axis 
br1 = np.arange(len(Calories)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
 
# Make the plot
plt.bar(br1, Calories, color ='r', width = barWidth, 
        edgecolor ='grey', label ='Calories') 
plt.bar(br2, CaloriesFat, color ='g', width = barWidth, 
        edgecolor ='grey', label ='CaloriesFat') 
plt.bar(br3, TotalFat, color ='b', width = barWidth, 
        edgecolor ='grey', label ='Totalfat') 
 
# Adding Xticks 
plt.xlabel('Item', fontweight ='bold', fontsize = 17) 
plt.ylabel('Nutrients', fontweight ='bold', fontsize = 17) 
plt.xticks([r + barWidth for r in range(len(Calories))], 
        ['Ckn Nuggts 40', 'BigBreakyHotChocLarge', 'Ckn Nuggts 20', 'BigBreakyHotChocRegular', 'BigBreakyLarge'])
 
plt.legend()
plt.show()
```
