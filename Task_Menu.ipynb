# Top 5 Calories & Fat
### After analyzing the menu's facts a pattern is found and this 5 items kept the same value for the 3 different attributes

```python
data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv")
# set width of bar 
barWidth = 0.25
fig = plt.subplots(figsize =(17, 8)) 
 
# set height of bar 
Calories = [1880, 1150, 1090, 1050, 990] 
CaloriesFat = [1060, 540, 530, 510, 470] 
TotalFat = [118, 60, 59, 56, 52] 
 
# Set position of bar on X axis 
br1 = np.arange(len(Calories)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
 
# Make the plot
plt.bar(br1, Calories, color ='r', width = barWidth, 
        edgecolor ='grey', label ='Calories') 
plt.bar(br2, CaloriesFat, color ='g', width = barWidth, 
        edgecolor ='grey', label ='CaloriesFat') 
plt.bar(br3, TotalFat, color ='b', width = barWidth, 
        edgecolor ='grey', label ='Totalfat') 
 
# Adding Xticks 
plt.xlabel('Item', fontweight ='bold', fontsize = 17) 
plt.ylabel('Nutrients', fontweight ='bold', fontsize = 17) 
plt.xticks([r + barWidth for r in range(len(Calories))], 
        ['Ckn Nuggts 40', 'BigBreakyHotCakeLarge', 'Ckn Nuggts 20', 'BigBreakyHotChocRegular', 'BigBreakyLarge'])
 
plt.legend()
plt.show()
```

# DENSITY OF THE ITEMS

### The figure below shows that items after Snacks & Side in the primary table is high in sugar

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.barplot(x='Item', y='Sugars', data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv"), palette='magma')
plt.title('Density of Sugars')
```

### The figure below shows that the majority of the items are low on vitamin A 

```python
sns.scatterplot(x='Item', y='Vitamin A (% Daily Value)', data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv"))
plt.title('Density of Vitamin A')
```

### The figure below shows that the majority of the items are low on vitamin C

```python
sns.scatterplot(x='Item', y='Vitamin C (% Daily Value)', data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv"))
plt.title('Density of Vitamin C')
```

### The figure below shows that the items in the 'Coffee & Tea' and 'Smoothies & Shakes' category is high in calcium

```python
sns.scatterplot(x='Item', y='Calcium (% Daily Value)', data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv"))
plt.title('Density of Calcium')
```

### The figure below shows that the items in the 'Breakfast', 'Beef & Pork' and 'Chicken & Fish' category is high in Iron

```python
sns.scatterplot(x='Item', y='Iron (% Daily Value)', data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv"))
plt.title('Density of Iron')
```

### The figure below shows that the items in the 'Beef & Pork' and 'Chicken & Fish' category is high in Protein

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.kdeplot(data['Protein'], fill=True)
plt.title('Density of Protein')
plt.show()
```

#### The figure below shows that the items in the 'Breakfast', 'Salads' and 'Chicken & Fish' category is high in Protein and Fiber

```python
f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)
sns.barplot(x='Protein', y='Dietary Fiber', data = pd.read_csv(r"C:\Users\lxy_l\Downloads\menu.csv"), palette='magma')
plt.title('Protein & Dietary Fiber')
```
